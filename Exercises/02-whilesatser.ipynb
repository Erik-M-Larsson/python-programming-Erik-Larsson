## 3. Guess number game

```python
import random as rnd
number = rnd.randint(1,100)
guess = -1
nr_guesses = 0

while guess != number:
    guess = int(input("Gissa ett tal"))
    print(guess)
    nr_guesses += 1
    
    if guess == number: 
        print("Grattis rätt svar!")
    else:
        if guess > number:
            print("Lägre")
        else:
            print("Högre")

    print(f"Antal gissningar {nr_guesses}\n")
```

## 4. Multiplication game

```python
play = True
points = 0

while play == True:
    x = rnd.randint(1, 10)
    y = rnd.randint(1, 10)
    prod = x * y

    anser = int(input(f"{x} * {y} = "))

    if anser == prod:
        print("Bra jobbat!")
        points += 1
        
    else:
        print(f"Fel! Rätt svar är {prod}")
    
    print(f"Poäng: {points}")

    inp = input("Spela igen [j/n]?")

    if inp == "j":
        play = True
    elif inp == "n":
        play = False

print("Tack för denna gång!")
```

## 5. Check convergence

```python

n = 0 
summa, summa_0 = 0, 1
err = 0.0000000000000001

while abs(summa - summa_0) > err:
    summa_0 = summa
    summa += 1/(2**n)
    print(n, summa)
    n += 1

print(n, summa)
```
